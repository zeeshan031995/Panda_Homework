```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["Student ID"].count()
total_budget = school_data["budget"].sum()
average_math = round(school_data_complete["math_score"].mean(),6)
average_reading = round(school_data_complete["reading_score"].mean(),5)

student_passing_math =school_data_complete.loc[school_data_complete["math_score"]> 70,:].count()["student_name"]
percent_passing_math = student_passing_math/float(total_students)*100

student_passing_reading = school_data_complete.loc[school_data_complete["reading_score"]> 70,:].count()["student_name"]
percent_passing_reading = student_passing_reading/float(total_students)*100

overall_passing_rate = round((percent_passing_math + percent_passing_reading)/2)

summary_Data = pd.DataFrame({"Total Schools":[total_schools],
                             "Total Students":[total_students],
                             "Total Budget":[total_budget],
                             "Average Math Score":[average_math],
                             "Average Reading Score":[average_reading],
                             "Percent Passing Math":[percent_passing_math],
                             "Percent Passing Reading":[percent_passing_reading],
                             "Percent Overall Passing Rate":[overall_passing_rate]})
                             
                           
District_summary = pd.DataFrame(summary_Data,columns=["Total Schools", "Total Students", "Total Budget", "Average Math Score",
                                                        "Average Reading Score", "Percent Passing Math", "Percent Passing Reading",
                                                        "Percent Overall Passing Rate"])
                             
District_summary
```

```python
school_Name=school_data.set_index(["school_name"])["type"]
student_perschool = school_data_complete["school_name"].value_counts()

total_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
percap_school_budget = total_school_budget/student_perschool

average_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
average_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

school_pass_math = school_data_complete[(school_data_complete["math_score"]> 70)]
school_pass_reading = school_data_complete[(school_data_complete["reading_score"]> 70)]

per_school_passmath = school_pass_math.groupby(["school_name"]).count()["student_name"]/student_perschool * 100
per_school_passread = school_pass_reading.groupby(["school_name"]).count()["student_name"]/student_perschool * 100
overall_pass = (per_school_passmath + per_school_passread)/ 2

summary_perschool = pd.DataFrame({"School Type":school_Name,
                             "Total Students":student_perschool,
                             "Total School Budget":total_school_budget,
                             "Per Student Budget":percap_school_budget,
                             "Average Math Score":average_school_math,
                             "Average Reading Score":average_school_reading,
                             "Percent Passing Math":per_school_passmath,
                             "Percent Passing Reading":per_school_passread,
                             "Percent Overall Passing Rate":overall_pass})


summary_perschool = pd.DataFrame(summary_perschool,columns=["School Type","Total Students","Total School Budget"
                                      "Per Student Budget","Average Math Score","Average Reading Score",
                                      "Percent Passing Math","Percent Passing Reading","Percent Overall Passing Rate"])



summary_perschool.head()
```

```python
school_spending_bins = [0,585,615,645,675]
group_names = ["<585","585-629","630-644","645-675"]

summary_perschool["Spending Ranges (Per Student)"] = pd.cut(percap_school_budget,school_spending_bins, labels=group_names)

math_spending_scores = summary_perschool.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
reading_spending_scores = summary_perschool.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
math_spending_pass = summary_perschool.groupby(["Spending Ranges (Per Student)"]).mean()["Percent Passing Math"]
reading_spending_pass = summary_perschool.groupby(["Spending Ranges (Per Student)"]).mean()["Percent Passing Reading"]
overall_passing_rate = (math_spending_scores + reading_spending_scores)/2

Spending_Summary_perschool = pd.DataFrame({"Average Math Score":math_spending_scores,
                                           "Average Reading Score":reading_spending_scores,
                                           "Percent Passing Math":math_spending_pass,
                                           "Percent Passing Reading":reading_spending_pass,
                                           "Overall Passing Rate":overall_passing_rate})
Spending_Summary_perschool = Spending_Summary_perschool[["Average Math Score","Average Reading Score",
                                                         "Percent Passing Math", "Percent Passing Reading",
                                                         "Overall Passing Rate"]]
                                                                              
Spending_Summary_perschool.head()                                                                              
                  
```
