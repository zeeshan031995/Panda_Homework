```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
player_total = len(purchase_data["SN"].unique())
player_count = pd.DataFrame({"Total Players":player_total},index = [0])
player_count.head()
```

```python
unique_itemcount = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
number_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

Purchase_Analysis = pd.DataFrame({"Number of Unique Items":unique_itemcount,"Average Price":average_price,
                                  "Number of Purchases":number_purchases,"Total Revenue":total_revenue}, index=[0])

Purchase_Analysis.head()
```

```python
group_gender = purchase_data.groupby("Gender")
unique_members = group_gender["SN"].nunique()
Percentof_Players = (unique_members/player_total)*100

Summaryof_Gender = pd.DataFrame({"Total COunt":unique_members,"Percentage of Players":Percentof_Players})

Summaryof_Gender.head()
```

```python
group_gender = purchase_data.groupby("Gender")
purchase_count = group_gender["Age"].count()
avg_Purchase = group_gender["Price"].mean()
total_Purchase = group_gender["Price"].sum()
avgPurchase_Price = (group_gender["Price"].sum()/unique_members)

Summary_datafile = pd.DataFrame({"Purchase Count":purchase_count, "Average Purchase Price":avg_Purchase,
                                "Total Purchase value":total_Purchase,"Avg Total Purchase per Person":
                                avgPurchase_Price})
Summary_datafile.head()
```

```python
age_group = [0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,999.99]
group_name = ["<10","10-14","14-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_group, labels=group_name)

purchase_data_agegroup = purchase_data.groupby("Age Group")
unique_members = purchase_data_agegroup["SN"].nunique()
percent_of_players = (unique_members/player_total)*100
age_demographics = pd.DataFrame({"Total count":unique_members,"Percentage of Players":percent_of_players})
age_demographics.index.name = None

age_demographics
```

```python
purchase_count = purchase_data_agegroup["Age"].count()
avg_Purchase = purchase_data_agegroup["Price"].mean()
total_Purchase = purchase_data_agegroup["Price"].sum()
avg_Person_Purchase = (purchase_data_agegroup["Price"].sum()/unique_members)

Summary_Analysis_Age = pd.DataFrame({"Purchase count":purchase_count, "Average Purchase Price":avg_Purchase, "Total Purchase Price":total_Purchase,
                                     "Average Total Purchase Per person":avg_Person_Purchase})

Summary_Analysis_Age
```

```python
purchase_data_SN= purchase_data.groupby("SN")

purchase_count = purchase_data_SN["Age"].count()
avg_Purchase = purchase_data_SN["Price"].mean()
total_Purchase = purchase_data_SN["Price"].sum()
total_Purchase.sort_values(ascending=False)
Top_Spender = pd.DataFrame({"Purchase count":purchase_count, "Average Purchase Price":avg_Purchase, "Total Purchase Price":total_Purchase})

Top_Spender.head()
```

```python
popular_df = purchase_data.loc[:,["Item ID","Item Name","Price"]]
popular_group = purchase_data.groupby(["Item ID","Item Name"])

purchase_count = popular_group["Age"].count()
avg_Purchase = popular_group["Price"].mean()
total_Purchase = popular_group["Price"].sum()

Most_Popular = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_Purchase,
                             "Total Purchase Value":total_Purchase})

Most_Popular = Most_Popular.sort_values("Purchase count", ascending = False)

Most_Popular.head()

```
